5 SIMPLE LINEAR REGRESSION

```python

X = Tessa['Age'] #independent vaiable
y = Tessa['Height'] #dependent vaiable

#Add a constant independent vaiable
X = sm.add_constant(X)

# Regression model
model = sm.OLS(y, X).fit()

# the regression results
model.summary()
```
